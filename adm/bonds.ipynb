```python
from pathlib import Path
from datetime import datetime
import quant
```

# Connect to LSEG

```python
CONFIG_PATH = Path(str(__vsc_ipynb_file__)).parent.parent.joinpath("lseg-data.config.json")
session = quant.SessionProvider(CONFIG_PATH)
```

```python
with session as client:
    result = client.bonds.list_securities("NVDA")
    
    for bond in result:
        print(f"{bond.company()} - {bond.maturity()} - {bond.asset_class()} - {bond}")
        
    print("#"*25)
    result_l2 = client.bonds.upgrade_l1_bond(result)
    
    bond = result_l2[0]
    
    # for bond in result_l2[:1]:
    print(f"{bond.company()} - {bond.maturity()} - {bond.asset_class()} - {bond.coupon_rate}%")
    
    bond_pricing = client.bonds.history_df(bond, fields=["BID", "ASK"], interval=quant.Interval.DAY_7, start=datetime(2025, 4, 18), end=datetime(2026, 4, 18))
    # result_l2_df = client.bonds.upgrade_l1_bond_df(result)
```
