```python
from pathlib import Path
from datetime import datetime
import quant
```

```python
CONFIG_PATH = Path(str(__vsc_ipynb_file__)).parent.parent.joinpath("lseg-data.config.json")
session = quant.SessionProvider(CONFIG_PATH)
```

```python
with session as client:
    bonds = client.bonds.upgrade_l1_bond(client.bonds.list_securities("MSFT"))
    
    equity = next(filter(lambda x: x.ric == "MSFT.O", client.equities.list_securities("MSFT")))
    common_stock = client.equities.upgrade_l1_equity_df(equity, fields=["TR.IMPLIEDVOLATILITY"])


    equity_value = common_stock["Company Market Capitalization"].iloc[0]
    equity_vol = common_stock["Implied Volatility"].iloc[0]
```

```python
result = quant.kmv.run_kmv_for_issuer(
    equity_value=equity_value,   # same units as amount_outstanding
    equity_volatility=0.22,
    bonds=bonds,
    risk_free_rate=0.05,
    empirical_table=quant.kmv.EMPIRICAL_TABLE,
)
print(result)
```
